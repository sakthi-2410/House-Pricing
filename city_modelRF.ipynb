```python
y=df[['AQI_Bucket']]
y
```

```python
xtrain, xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2)
```

```python
clf = RandomForestClassifier()
clf.fit(xtrain,ytrain)
ypred = clf.predict(xtest)
score = accuracy_score(ytest,ypred)
print(score)
```
